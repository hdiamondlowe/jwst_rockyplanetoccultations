```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii
import astropy.units as u
from astropy.table import Table, vstack, unique
from astropy.time import Time
import astrotools.orbitparams as orb

import starry as st
st.config.lazy = False
st.config.quiet = True
```

```python
# Creating a standard star
def make_star(targ):
    M_s  = targ['st_mass']
    R_s  = targ['st_rad']
    prot = 1

    star = st.Primary(
        st.Map(ydeg = 1, udeg = 2, nw = 1, amp = 1.0), 
        m    = M_s,
        r    = R_s, 
        prot = prot
    )

    return star

# creating a planet; 
# adapted from Mette's code
def make_planet(targ, tidally_locked=True):

    planet  = st.kepler.Secondary(
        st.Map(ydeg = 5, nw = 1, amp = 1e-3),               
        m      = targ['pl_bmasse'],
        r      = targ['pl_rade'],
        porb   = targ['pl_orbper'],                                                      
        prot   = targ['pl_orbper'],                                       
        Omega  = 0,                                         
        ecc    = 0,                                                         
        w      = 90,                                  
        t0     = targ['pl_tranmid'],                           
        theta0 = 0,                           
        inc    = targ['pl_orbincl'],
        length_unit = u.Rearth,
        mass_unit   = u.Mearth

    )
    
    if tidally_locked:
        planet.map.spot(contrast = -1, radius = 60)
        
    return planet
```

```python
def get_bad_occultation(star):
    
    starname = star['hostname']
    print(f'Making {starname} system')

    # retrieve all the planets around this star
    inds = np.argwhere(star['hostname'] == samplePlanets['hostname'])
    planets = []
    for i in inds: planets.append(samplePlanets[i])
    planets = vstack(planets)
    
    
    # make the star and planets in starry
    starry_star = make_star(star)
    starry_other_planets = []
    
    # get the index of the "target planet" -- the one in the sample
    target_planet_ind = 0
    for i, pl in enumerate(planets): 
        if pl['pl_name'] in sample['pl_name']: 
            #print(pl['pl_name'], sample['pl_name'])
            print('Target planet index = ', i)
            target_planet_ind += i
            starry_target_planet = make_planet(pl, tidally_locked=False)
            
        else: starry_other_planets.append(make_planet(pl, tidally_locked=False))

    print('Total number of planets = ', len(starry_planets))
    
    # cannot figure out how to add an arbitrary number of planets so here we are
    if len(planets)==1: system = st.System(starry_star, starry_target_planet)
    elif len(planets)==2: system = st.System(starry_star, starry_target_planet, starry_other_planets[0])
    elif len(planets)==3: system = st.System(starry_star, starry_target_planet, starry_other_planets[0], starry_other_planets[1])
    elif len(planets)==4: system = st.System(starry_star, starry_target_planet, starry_other_planets[0], starry_other_planets[1], starry_other_planets[2])
    else:
        print('ERROR: You need to add more planet options! (Or figure out a better way to do this)')
        return
    
    # get all of the occultation times in the observing period for the target planet
    target_planet = planets[target_planet_ind]
    target_planet_name = target_planet['pl_name']
    print(f'Finding the bad occultations for {target_planet_name}')

    tdur = orb.Tdur(P=target_planet['pl_orbper']*u.day, 
                    Rp_Rs=((target_planet['pl_rade']*u.R_earth)/(target_planet['st_rad']*u.R_sun)).decompose(),
                    a_Rs = ((target_planet['pl_orbsmax']*u.AU)/(target_planet['st_rad']*u.R_sun)).decompose(),
                    i = target_planet['pl_orbincl']
                    ) # event duration

    start_time = Time("2023-07-01T00:00:00").jd
    end_time   = Time("2025-01-01T00:00:00").jd

    T0 = Time(target_planet['pl_tranmid'], format='jd', scale='utc').value
    start_num = int(np.round((start_time - T0)/target_planet['pl_orbper']))
    end_num   = int(np.round((end_time - T0)/target_planet['pl_orbper']))
    event_num = np.arange(start_num, end_num+1)

    t0p5s = target_planet['pl_tranmid'] + (event_num+0.5)*target_planet['pl_orbper']
    
    # get the transit/occultation events of the other planets in the system
    other_events = []
    for i, other_planet in enumerate(planets):
        if i == target_planet_ind: continue

        T0 = Time(other_planet['pl_tranmid'], format='jd', scale='utc').value
        start_num = int(np.round((start_time - T0)/other_planet['pl_orbper']))
        end_num   = int(np.round((end_time - T0)/other_planet['pl_orbper']))
        event_num = np.arange(start_num, end_num+1)

        transits = other_planet['pl_tranmid'] + event_num*other_planet['pl_orbper']
        eclipses = other_planet['pl_tranmid'] + (event_num+0.5)*other_planet['pl_orbper']
        
        other_events.append(transits)
        other_events.append(eclipses)

    other_events = np.sort(np.hstack(other_events))
    
    # find the overlapping events with the target occultations
    n_bad_events = 0
    for t_occ in t0p5s:
        t_start = t_occ - 3*tdur.to(u.day).value
        t_end   = t_occ + 3*tdur.to(u.day).value
        #t_start = -3*tdur.to(u.day).value
        #t_end   =  3*tdur.to(u.day).value
        time = np.linspace(t_start, t_end, 1000)

        # get closest event to occulation
        event_ind = np.argmin(abs(t_occ - other_events))
        closest_event = other_events[event_ind]

        #if closest_event >= t_start and closest_event <= t_end:

        flux = np.hstack(system.flux(time))

        plt.plot(time, flux)
        plt.show()

        print('Do not observe between:', t_start, 'and', t_end)
        n_bad_events += 1
            
    print(f'Number of bad occultations = for {target_planet_name}', n_bad_events, 'out of', len(t0p5s))
```
