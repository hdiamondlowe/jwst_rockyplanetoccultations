```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigf_transit(xf, xn, L, sigL):
    varf = (sigL/L)**2 * ( 1/(xf**3 * xn**3) + 1/(xf**2 * xn**2) )
    return np.sqrt(varf)
```

```python
def sigf_eclipse(xf, xn, L, sigL):
    varf = (sigL/L)**2 * ( xf*xn + xf**2*xn**2 )
    return np.sqrt(varf)
```

```python
# for transit
L = 0.998
sigL = 0.0005
xn = np.linspace(0.8, 5, 100)
xf = 1/L

sigf = sigf_transit(xf, xn, L, sigL)
plt.plot(xn, sigf)

plt.axvline(1, color='k')

plt.xlabel('baseline observine/in-transit observint')
plt.ylabel('err in transit depth')
```

```python
# for eclipse
L = 1
sigL = 0.0005
xn = np.linspace(0.001, 5, 100)
xf = 1.0005

sigf = sigf_eclipse(xf, xn, L, sigL)
plt.plot(xn, sigf)

plt.axvline(1, color='k')

plt.xlabel('baseline observine/in-transit observint')
plt.ylabel('err in eclipse depth')
```

```python
def sigf_from_PandExo(L, sigL, nH, nL):
    sigf = sigL / L * (1/np.sqrt(nH)) * np.sqrt(nH/nL + 1)
    return sigf
```

```python
# for whatever
L = 1
sigL = 0.001
nL = 1
nH = np.linspace(1, 5, 100) * nL


sigf = sigf_from_PandExo(L, sigL, nH, nL)
plt.plot(nH, sigf)

plt.axhline(sigf_from_PandExo(L, sigL, 2*nL, nL), color='k')

print('test', sigf_from_PandExo(L, sigL, nL, nL)/sigf_from_PandExo(L, sigL, 2*nL, nL))
print(np.sqrt(2/1.5))

plt.xlabel('baseline observine/in-transit observint')
plt.ylabel('err in eclipse depth')
```
